```python
#Put the 75 clusters in the top200
import pandas as pd
data_preds = pd.read_csv('datawithpreds.csv')
data_preds.head()
```

```python
Top200 = pd.read_csv('top200features.csv')
Top200.head()
```

```python
import numpy as np
numeric = Top200.select_dtypes(include=[np.number])
numeric.head()
```

```python
numeric[np.isnan(numeric)] = 0  #dealing with nans
numeric.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

Normalized=numeric
scaler = MinMaxScaler()

for col in numeric.columns:
    Normalized[col] = scaler.fit_transform(numeric[[col]])

Normalized.head()
```

```python
x_prep=numeric
x_prep.head()
```

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors

k = len(x_prep.columns) + 1 
nearest_neighbors = NearestNeighbors(n_neighbors=k)
neighbors=nearest_neighbors.fit(x_prep)
distances,indices=neighbors.kneighbors(x_prep)
distances = np.sort(distances[:,-1],axis=0)
```

```python
from sklearn.cluster import DBSCAN

dbscan=DBSCAN(eps=0.5,min_samples=1) #these values were defined by evalutaion made in lab26
clusters=dbscan.fit_predict(x_prep)

x_prep["clusters"]=clusters
x_prep["clusters"].value_counts().sort_index()
```

```python
topclusters = {
    "title": Top200["title"],
    "clusters": x_prep["clusters"]
}
```

```python
print(topclusters)
```

```python
import pickle

topclusters = pd.DataFrame(topclusters)  

pickle_file = 'top200.pickle'
with open(pickle_file, 'wb') as f:
    pickle.dump(topclusters, f)

print(f"Pickle file '{pickle_file}' created successfully.")
```
